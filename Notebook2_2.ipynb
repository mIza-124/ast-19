```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 1, 100)
```

```python
def f_sin(x):
    return np.sin(x)

def f_cos(x):
    return np.cos(x)
```

```python
y_sin = f_sin(x)
y_cos = f_cos(x)

fig, ax = plt.subplots(1, 2)

ax[0].plot(x, y_sin)
ax[0].set_title("sin(x)")

ax[1].plot(x, y_cos)
ax[1].set_title("cos(x)")

plt.savefig("trig_plots.pdf")
plt.show()
```
