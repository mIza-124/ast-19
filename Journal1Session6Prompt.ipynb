# Trigonometric Functions

This notebook defines functions for sin(x) and cos(x), tabulates them over a range, and prints the initial values.

```python
import numpy as np
```

Defines function for sin(x)

```python
def get_sin(x):
    return np.sin(x)
```

Defines function for cos(x)

```python
def get_cos(x):
    return np.cos(x)
```

Creates a lot with values between 0 and 2 with 1000 entries

```python
x_values = np.linspace(0, 2, 1000)
```

get sin(x) and cos(x) for x values

```python
sin_values = get_sin(x_values)
cos_values = get_cos(x_values)
```

Prints header for the table first then loops through the first 10 indicies and prints values below

```python
print(f"{'x':<10} {'sin(x)':<10} {'cos(x)':<10}")
print("-" * 30)

for i in range(10):
    print(f"{x_values[i]:<10.4f} {sin_values[i]:<10.4f} {cos_values[i]:<10.4f}")
```
